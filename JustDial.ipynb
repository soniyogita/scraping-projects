```python
from selenium import webdriver  
from selenium.webdriver.common.keys import Keys  
from selenium.webdriver.chrome.options import Options 
import time
from bs4 import BeautifulSoup
import pandas as pd
import csv
SCROLL_PAUSE_TIME = 10



head=["Name","Category","city","Address","Email"]
with open('just_dial.csv', mode='w+',encoding="UTF-8",newline="") as file:
    writer1 = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    writer1.writerow(head)
    City_Name="jodhpur"                                                                  #city Name insertion
    Category="schools"                                                                   #category insertion
    chrome_options = Options()  
    chrome_options.add_argument("--headless")  
    driver =webdriver.Chrome(executable_path=r'C:\Users\yogita\Desktop\chromedriver_win32\chromedriver.exe')
    url="https://www.justdial.com/"+City_Name+"/"+Category
    driver.get(url)
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    for i in range(5):
        driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")      #for scroll down the page
        time.sleep(SCROLL_PAUSE_TIME)
    time.sleep(10)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    names=soup.find_all("h2",{"class":"store-name"})
    for name in names:
        add=name.find("a")
        a=add.get("href")
        print(a)
        driver.get(a)
        soup1=BeautifulSoup(driver.page_source, 'html.parser')
        All_list=[]
        all_items=soup1.find("ul",{"class":"comp-contact"})
        if all_items.find("i",{"class":"web_ic sprite_wb comp-icon"}) is not None:
            www=all_items.find_all("a")
            for unit_a in www:
                if "www." in unit_a.text:
                    emai=unit_a.get("href")
                    if "http:" in emai:
                        email=emai.strip("http://")                         #find Email of the company
                        print(email)
                    elif "https:" in emai:
                        email=emai.strip("https://")
                    All_list.append(email)
                elif ".in" in unit_a.text:
                    email=unit_a.get("href")
            if all_items.find("span",{"class":"lng_add"}) is not None:
                address=all_items.find("span",{"class":"lng_add"}).text             #find address 
                #print(address)
                All_list.append(address)
            else:
                All_list.append("None")
            Name=soup1.find("span",{"class":"item"}).text
            #print(Name)
            All_list.append(Name)
            print(All_list)
            writer1.writerow([Name,Category,City_Name,address,email])           #insert in the csv file 

```
